```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

without = pkl.load(open("Cont_PG.pkl", "rb"))
fixed = pkl.load(open("Cont_PG_fixed_bl.pkl", "rb"))
value = pkl.load(open("Cont_PG_value_bl_scaled.pkl", "rb"))
value_lucky = pkl.load(open("Cont_PG_value_bl_scaled_lucky.pkl", "rb"))
advantage = pkl.load(open("Cont_PG_value_bl_scaled_advantage_store_advantage.pkl", "rb"))


without = np.array(without) 
fixed = np.array(fixed)
value = np.array(value)
value_lucky = np.array(value_lucky)
advantage = np.array(advantage)

# value = (value+value_lucky)/2

K = 50

plt.figure()

# plt.plot(without )
plt.plot(np.convolve(without, np.ones(K)/K ) )
plt.title("Without Baseline")
plt.xlabel("Episodes")
plt.ylabel("Rewards")
# plt.show()

# plt.plot(fixed )
plt.plot(np.convolve(fixed, np.ones(K)/K ) )
plt.title("State Independent Baseline")
plt.xlabel("Episodes")
plt.ylabel("Rewards")
# plt.show()

# plt.plot(value)
plt.plot(np.convolve(value, np.ones(K)/K ) )
plt.title("Value Function Baseline")
plt.xlabel("Episodes")
plt.ylabel("Rewards")

# plt.plot(advantage)
# plt.plot(np.convolve(advantage, np.ones(K)/K ) )
# plt.title("Advantage of reward after scaling")
# plt.xlabel("Episodes")
# plt.ylabel("Advantage")

plt.title("InvertedPendulum-v2")
plt.legend(["Without Baseline", "State Independent Baseline", "Value Function Baseline"])
plt.show()

```
