```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

without = pkl.load(open("PG.pkl", "rb"))
fixed = pkl.load(open("PG_fixed_bl.pkl", "rb"))
value = pkl.load(open("PG_value_bl.pkl", "rb"))
value_scale = pkl.load(open("PG_value_bl_scaled.pkl", "rb"))


without = np.array(without).mean(0) 
fixed = np.array(fixed).mean(0)
value = np.array(value).mean(0)
value_scale = np.array(value_scale).mean(0)

K = 10

plt.figure()

# plt.plot(without )
plt.plot(np.convolve(without, np.ones(K)/K ) )
# plt.title("Without Baseline")
plt.xlabel("Episodes")
plt.ylabel("Rewards")
# plt.show()

# plt.plot(fixed )
plt.plot(np.convolve(fixed, np.ones(K)/K ) )
# plt.title("State Independent Baseline")
plt.xlabel("Episodes")
plt.ylabel("Rewards")
# plt.show()

# plt.plot(value)
plt.plot(np.convolve(value, np.ones(K)/K ) )
# plt.title("Value Function Baseline")
plt.xlabel("Episodes")
plt.ylabel("Rewards")
# plt.show()

# plt.plot(value_scale )
plt.plot(np.convolve(value_scale, np.ones(K)/K ) )
# plt.title("Value Function Baseline Scaled Reward")
plt.xlabel("Episodes")
plt.ylabel("Rewards")

plt.legend(["Without Baseline", "State Independent Baseline", "Value Function Baseline", "Value Function Baseline Scaled Reward"])
plt.show()
```
